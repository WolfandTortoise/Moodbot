```python
import random

class MoodBot:
    def __init__(self):
        self.mood_phrases = {
            "positive": ["happy", "excited", "joyful", "content", "positive", "good"],
            "negative": ["sad", "angry", "stressed", "anxious", "negative", "not good"]
        }

        self.response_phrases = {
            "What is the weather": ["weather"],
            "What are you doing": ["doing"]
        }

    def detect_mood(self, user_input):
        user_input = user_input.lower()
        for mood, phrases in self.mood_phrases.items():
            if any(phrase in user_input for phrase in phrases):
                return mood
        return "neutral"

    def suggest_activity(self, mood):
        if mood == "positive":
            return "Great to hear that you're feeling positive! How about going for a walk or doing something you enjoy?"
        elif mood == "negative":
            return "I'm sorry to hear that you're feeling down. Consider trying a mindfulness exercise or talking to a friend."
        else:
            return "How else can I help you today?"

    def detect_response(self, user_input):
        user_input = user_input.lower()
        for response, phrases in self.response_phrases.items():
            if any(phrase in user_input for phrase in phrases):
                return response

        return "neutral"

    def chat_with_user(self):
        print("Hello! I'm MoodBot. How are you feeling today?")
        while True:
            user_input = input("User: ")
            if user_input.lower() in ["exit", "bye", "quit"]:
                print("Goodbye! Take care.")
                break

            mood = self.detect_mood(user_input)
            response = self.suggest_activity(mood)
            detected_response = self.detect_response(user_input)

            if detected_response == "What is the weather":
                print("MoodBot: The weather in Tallinn is 5 degrees Celsius.")
            elif detected_response == "What are you doing":
                print("MoodBot: Nothing much, I am here to help you.")
            else:
                print("MoodBot:", response)

if __name__ == "__main__":
    mood_bot = MoodBot()
    mood_bot.chat_with_user()
```
